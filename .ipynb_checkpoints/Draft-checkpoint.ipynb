```python
import os
import numpy
import cv2
import shutil
```

```python
main_dir_picts = 'Data/Images'
l_imgs = os.listdir(main_dir_picts)
```

```python
im = cv2.imread(os.path.join(main_dir_picts, l_imgs[0]))
```

```python
im.shape
```

```python
os.mkdir('Data/ImagesTiles')
```

```python
("{}_{}_{}.jpg").format(l_imgs[0].split('.')[0], 9, 3)
```

```python
os.mkdir(os.path.join('Data/ImagesTiles', l_imgs[0].split('.')[0]))
```

```python
lx = list(range(0,im.shape[0], 512))
ly = list(range(0, im.shape[1], 512))
lastx = lx[-1] - (512 - im.shape[0] % 512)
lasty = ly[-1] - (512 - im.shape[1] % 512)
lx.append(lastx)
ly.append(lasty)


for xi in range(len(lx)):
    for yi in range(len(ly)):
        #print('xi ', xi, ' yi ', yi)
        if xi != len(lx) -2 and yi != len(ly) -2:
            print('xi ', xi, ' yi ', yi)
            if xi <= len(lx) -3 and yi <= len(ly) -3:
                xmin = lx[xi]
                xmax = lx[xi+1]
                ymin = ly[yi]
                ymax = ly[yi+1]
                ctile = im[xmin:xmax, ymin:ymax, :]
                cpath = os.path.join('Data/ImagesTiles', l_imgs[0].split('.')[0])
                cv2.imwrite( os.path.join( cpath,\
                                          ("{}_{}_{}.jpg").format(l_imgs[0].split('.')[0], xmin, ymin )), ctile)
            elif xi == len(lx) -1  and yi <= len(ly) - 3:
                xmin = lx[xi]
                xmax = im.shape[0]
                ymin = ly[yi]
                ymax = ly[yi+1]
                ctile = im[xmin:xmax, ymin:ymax, :]
                cpath = os.path.join('Data/ImagesTiles', l_imgs[0].split('.')[0])
                cv2.imwrite( os.path.join( cpath,\
                                          ("{}_{}_{}.jpg").format(l_imgs[0].split('.')[0], xmin, ymin )), ctile)

            elif xi <= len(lx) - 3 and yi == len(ly) -1 :
                xmin = lx[xi]
                xmax = lx[xi+1]
                ymin = ly[yi]
                ymax = im.shape[1]
                ctile = im[xmin:xmax, ymin:ymax, :]
                cpath = os.path.join('Data/ImagesTiles', l_imgs[0].split('.')[0])
                cv2.imwrite( os.path.join( cpath,\
                                          ("{}_{}_{}.jpg").format(l_imgs[0].split('.')[0], xmin, ymin )), ctile)

            elif  xi == len(lx) -1 and yi == len(ly) -1:
                xmin = lx[xi]
                xmax = im.shape[0]
                ymin = ly[yi]
                ymax = im.shape[1]
                ctile = im[xmin:xmax, ymin:ymax, :]
                cpath = os.path.join('Data/ImagesTiles', l_imgs[0].split('.')[0])
                cv2.imwrite( os.path.join( cpath,\
                                          ("{}_{}_{}.jpg").format(l_imgs[0].split('.')[0], xmin, ymin )),ctile)
        else:
            print(" Skip {} {} ".format(lx[xi], ly[yi]), xi , yi)
```

##  Format folder

```python
os.mkdir('JPEGImage')
```

```python
os.mkdir('SegmentationClass')
```

```python
main_dir_picts = 'Data/TiledImages'
l_main_dir_picts = os.listdir(main_dir_picts)
for folder in l_main_dir_picts:
    img_l = os.listdir(os.path.join(main_dir_picts, folder))
    for im in img_l:
        shutil.copy(os.path.join(main_dir_picts, folder, im), os.path.join('JPEGImage', im))
```

```python
from PIL import Image
main_dir_picts = 'Data/TiledMasks'
l_main_dir_picts = os.listdir(main_dir_picts)
for folder in l_main_dir_picts:
    img_l = os.listdir(os.path.join(main_dir_picts, folder))
    for im in img_l:
        n_name = str(str(im.split('-labels')[0] + im.split('-labels')[1]).split('jpg')[0]+'png')
        print(n_name)
        im1 = Image.open(os.path.join(main_dir_picts, folder, im))
        im1.save('SegmentationClass/'+ n_name)
```
