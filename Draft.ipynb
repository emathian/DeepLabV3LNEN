```python
main_dir_picts = 'Data/TiledImages'
l_main_dir_picts = os.listdir(main_dir_picts)
for folder in l_main_dir_picts:
    img_l = os.listdir(os.path.join(main_dir_picts, folder))
    for im in img_l:
        shutil.copy(os.path.join(main_dir_picts, folder, im), os.path.join(path_to_main_folder + 'TumorDetection/dataset/JPEGImages', im))
```

```python
from PIL import Image
main_dir_picts = 'Data/TiledMasks'
l_main_dir_picts = os.listdir(main_dir_picts)
for folder in l_main_dir_picts:
    img_l = os.listdir(os.path.join(main_dir_picts, folder))
    for im in img_l:
        n_name = str(str(im.split('-lab1els')[0] + im.split('-labels')[1]).split('jpg')[0]+'png')
        im1 = Image.open(os.path.join(main_dir_picts, folder, im))
        im1.save(path_to_main_folder + 'TumorDetection/dataset/SegmentationClass/' + n_name)
```

```python
l_all_picts = os.listdir(path_to_main_folder + 'TumorDetection/dataset/JPEGImages')
travail = []
train = []
test = []
for pict in l_all_picts:
    fname = pict.split('.')[0]
    travail.append(fname)

    if fname.find('TNE0887') == -1 and fname.find('TNE0970') == -1:
        train.append(fname)
    else:
        test.append(fname)
    
```
